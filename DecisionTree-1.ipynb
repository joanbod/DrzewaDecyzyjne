Drzewa decyzyjne są graficzną metodą wspomagania procesu decyzyjnego. Jest to jedna z technik analizy danych. Drzewo składa się z korzenia oraz gałęzi prowadzących z korzenia do kolejnych wierzchołków. Wierzchołki, z których wychodzi co najmniej jedna krawędź, są nazywane węzłami, a pozostałe wierzchołki – liśćmi. W każdym węźle sprawdzany jest pewien warunek dotyczący konkretnej obserwacji i na jego podstawie wybierana jest jedna z gałęzi prowadząca do kolejnego wierzchołka. Klasyfikacja danej obserwacji polega na przejściu od korzenia do liści i przypisaniu do tej obserwacji klasy zapisanej w danym liściu. W eksploracji danych i uczeniu maszynowym drzewa decyzyjne są modelami prognozującymi.

Używane są w kilku przypadkach:
- redukcji ilości danych poprzez przekształcenie ich do formy zajmującej mniej „miejsca” przy jednoczesnym zachowaniu cech oraz zapewnieniu dokładności rozwiązania problemu;
- sprawdzania, czy dane zawierają oddzielne grupy obiektów, takie że te grupy można sensownie interpretować pod kątem teorii;
- wykrycia mapowania z niezależnych do zależnych zmiennych, które jest użyteczne przy przewidywaniu wartości zależnej zmiennej w przyszłości.

### Zastosowanie drzew decyzyjnych
Drzewa decyzyjne znajdują praktyczne zastosowanie w różnego rodzaju problemach decyzyjnych, szczególnie takich, gdzie występuje dużo rozgałęziających się wariantów, a także w warunkach ryzyka. 
Drzewa decyzyjne znajdują szerokie zastosowanie w :
- diagnostyce medycznej,
- przewidywaniu wydajności,
- akceptacji i udzielaniu kredytów.

Algorytm drzewa decyzyjnego to algorytm uczenia nadzorowanego, który może być używany zarówno do zadań klasyfikacji, jak i regresji. Podstawową ideą drzewa decyzyjnego jest rekurencyjny podział przestrzeni wejściowej (tj. zestawu możliwych instancji wejściowych) na coraz mniejsze regiony, zwane węzłami liści, w oparciu o wartości cech wejściowych. Celem drzewa decyzyjnego jest przewidywanie etykiety klasy lub wartości nowej instancji na podstawie jej cech.

Dostępnych jest kilka algorytmów służących do klasyfikacji i analizy segmentacji. Wszystkie te algorytmy zasadniczo realizują to samo zadanie: dzieląc dane na kolejne podgrupy, analizują wszystkie zmienne w zbiorze danych, by znaleźć zmienną zapewniającą najlepszą klasyfikację lub predykcję. Proces jest rekursywny, a grupy są dzielone na coraz mniejsze jednostki aż do ukończenia drzewa (zgodnie z określonym kryterium zatrzymania). Zmienne przewidywane i wejściowe używane do budowania drzewa mogą być ilościowe (przedział liczbowy) lub jakościowe, w zależności od algorytmu. Jeśli zmienna przewidywana jest ilościowa, generowane jest drzewo regresji; jeśli zmienna przewidywana jest jakościowa, generowane jest drzewo klasyfikacji.



### Działanie algorytmu drzewa decyzyjnego:

- **Wybierz najlepszą funkcję do podziału danych:** Algorytm rozpoczyna się od wybrania funkcji, która najlepiej rozdziela dane na różne klasy lub wartości. Ta funkcja jest nazywana węzłem głównym drzewa.

- **Podziel dane na podstawie wybranej cechy:** Algorytm następnie dzieli dane na podzbiory na podstawie wartości wybranej cechy. Każdy podzbiór odpowiada gałęzi drzewa.

- **Utwórz węzły wewnętrzne:** Algorytm tworzy nowy węzeł wewnętrzny dla każdego podzbioru i przypisuje funkcję używaną do podziału jako atrybut węzła.

- **Utwórz węzły liścia:** jeśli podzbiór danych jest czysty, co oznacza, że ​​wszystkie instancje w podzbiorze należą do tej samej klasy lub mają tę samą wartość, algorytm tworzy węzeł liścia dla tego podzbioru i przypisuje etykietę klasy lub wartość do węzeł.

- **Powtórz proces dla każdego podzbioru:** Algorytm powtarza proces wybierania najlepszej cechy, dzielenia danych, tworzenia węzłów wewnętrznych i tworzenia węzłów liścia dla każdego podzbioru, aż do spełnienia kryterium zatrzymania. Kryterium zatrzymania może być oparte na liczbie wystąpień w podzbiorze, liczbie cech lub czystości podzbiorów.

- **Użyj drzewa do przewidywania:** po zbudowaniu drzewa można go użyć do przewidywania etykiety klasy lub wartości nowej instancji, przechodząc przez drzewo od korzenia do węzła liścia. Etykieta klasy lub wartość węzła liścia to przewidywana etykieta klasy lub wartość instancji.



## Przykład drzewa decyzyjnego
Rozpoczynamy importem niezbędnych bibliotek dla klasyfikatora drzewa decyzyjnego, zestawu danych irysów, funkcji train_test_split, metody matplotlib i plot_tree z scikit-learn. Następnie ładujemy zestaw danych irysów, który składa się ze 150 próbek kwiatów irysów, z których każda ma cztery cechy (długość działki kwiatu, szerokość działki kwiatu, długość płatka i szerokość płatka) oraz etykietę (0, 1 lub 2, reprezentującą gatunek irysa).

Następnie podzieliłyśmy dane na zestawy treningowe i testowe za pomocą funkcji train_test_split o rozmiarze testu 0,25, co oznacza, że ​​25% danych zostanie użytych do testowania, a pozostałe 75% do treningu.

Kolejno tworzymy instancję klasy DecisionTreeClassifier, która jest klasyfikatorem. Dopasowujemy klasyfikator do danych treningowych za pomocą metody `fit`. Używamy klasyfikatora do przewidywania etykiet danych testowych za pomocą metody przewidywania i obliczania dokładności prognoz. 

```python
# Import the necessary libraries
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Load the iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Create a decision tree classifier
clf = DecisionTreeClassifier()

# Train the classifier using the training data
clf.fit(X_train, y_train)

# Test the classifier on the test data
accuracy = clf.score(X_test, y_test)
print("Accuracy: ", accuracy)

# Plot the decision tree
plt.figure(figsize=(15,7))
plot_tree(clf, filled=True, feature_names=iris.feature_names,class_names=iris.target_names, node_ids=True,proportion=True)
```

Dzięki bibliotece matplotlib wykreślamy drzewo decyzyjne za pomocą metody plot_tree. Uzyskujemy graficzną reprezentację drzewa decyzyjnego, która może pomóc w zrozumieniu, w jaki sposób klasyfikator dokonuje prognoz na podstawie cech danych.

Dodatkowo możemy edytować parametry klasyfikatora drzewa decyzyjnego, aby zoptymalizować wydajność modelu, np. ustawienie parametru `max_depth`, aby ograniczyć głębokość drzewa lub ustawienie parametru `min_samples_split`, aby określić minimalną liczbę próbek wymaganych do podziału węzła wewnętrznego.

Można również użyć innych technik, takich jak przycinanie, walidacja krzyżowa i metody zespołowe, aby uniknąć nadmiernego dopasowania i poprawić uogólnienie modelu.

### Podsumowanie
Drzewa decyzyjne nie są najefektywniejszymi modelami w uczeniu maszynowym. Jednakże mają kilka zalet które powodują, warto  z nich korzystać:
- łatwe w zrozumieniu i interpretacji 
- możliwość używania danych ilościowych i jakościowych – jeśli mamy dane ilościowe (np. wzrost – 1.88, 1,90 itd.) to funkcja modelująca będzie szukać optymalnej wartości do podziału. W przypadku danych jakościowych (dobry, lepszy, najlepszy) wystarczy je odpowiednio zakodować.
- dane nie wymagają dużych przygotowań  

Drzewa decyzyjne mają też kilka wad:
- utworzone drzewa mogą być skomplikowane – przy dużej ilości obserwacji i cech drzewa które powstaną będą miały dużo poziomów i bardzo dużo liści.
- są zmienne – dodanie albo odjęcie nawet jednej obserwacji może spowodować zbudowanie całkiem innego drzewa.
- mają dość dużą złożoność obliczeniową – funkcja modelująca musi sprawdzić wszystkie możliwe wartości dla wszystkich cech.
